```python
Url = "https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv"

covid = pd.read_csv(Url)
```

# 3. Low Level Data Understanding

## a) count of unique values in location column. 

```python
len(pd.unique(covid['location'])) 
```

## b) continent that has maximum frequency

```python
max_freq = covid['continent'].value_counts()
print("continent that has maximum frequency: ",max_freq[[0]])
```

## c) maximum & mean value in 'total_cases'. 

```python
print("maximum value in 'total_cases': ", covid['total_cases'].max())
print("mean value in 'total_cases': ", covid['total_cases'].mean())
```

## d) 25%,50% & 75% quartile value in 'total_deaths' 

```python
print(covid.total_deaths.quantile([0.25,0.5,0.75]))
```

## e) continent that has maximum 'human_development_index'. 

```python
print("continent that has maximum 'human_development_index':", covid[covid['human_development_index'] == covid['human_development_index'].max()].continent.unique())
```

## f) continent that has minimum 'gdp_per_capita'. 

```python
print("continent that has maximum 'gdp_per_capita':", covid[covid['gdp_per_capita'] == covid['gdp_per_capita'].min()].continent.unique())
```

# 4. Filter the dataframe with only this columns

### ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] 

```python
covid = covid.filter(['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
```

```python
covid
```

# 5. Data Cleaning

## a. Remove all duplicates observations

```python
covid.drop_duplicates(keep=False,inplace=True)
covid
```

## b. Find missing values in all columns 

```python
covid.isnull()
```

## c. Remove all observations where continent column value is missing

```python
covid = covid[covid.continent != 0]
```

## d. Fill all missing values with 0 

```python
covid.fillna(0, inplace = True)
```

```python
covid.isnull()
```

# 6. Date time format

## a. Convert date column in datetime format using pandas.to_datetime

```python
covid["date"]= pd.to_datetime(covid["date"])
```

```python
covid.info()
```

## b. Create new column month after extracting month data from date column.

```python
covid['month']= covid['date'].dt.month 
```

```python
covid
```

# 7. Data Aggregation

## a. Find max value in all columns using groupby function on 'continent' column 

```python
covid.groupby(['continent']).max().reset_index()
```

## b. Store the result in a new dataframe named 'df_groupby' 

```python
df_groupby = covid.groupby(['continent']).max().reset_index()
```

```python
df_groupby
```

# 8. Feature Engineering

### a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases' 

```python
df_groupby['total_deaths_to_total_cases']=df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby
```

# 9. Data Visualization

### a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot. 

```python
displot = sns.displot(df_groupby['gdp_per_capita'])
displot.savefig("9a_displot.png")
```

## b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita' 

```python
scatterplot = sns.scatterplot(data=df_groupby, x='total_cases', y='gdp_per_capita')
fig = scatterplot.get_figure()
fig.savefig("9b_scatterplot.png")
```

## c. Plot Pairplot on df_groupby dataset. 

```python
fig = sns.pairplot(df_groupby)
fig.savefig("9c_pairplot.png")
```

## d. Plot a bar plot of 'continent' column with 'total_cases' .

```python
fig = sns.catplot(x='continent',y='total_cases',data=df_groupby,kind="bar")
fig.savefig("9d_barplot.png")
```
